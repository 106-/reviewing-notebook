```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# データ生成
np.random.seed(0)
x = 2.5 * np.random.randn(100) + 1.5   # 平均1.5、標準偏差2.5の正規分布から100個のデータを生成
res = 0.5 * np.random.randn(100)       # ノイズ項を生成
y = 2 + 0.3 * x + res                  # 真の関係は y = 2 + 0.3x + ノイズ

# データをプロット
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Input data')
plt.title("Generated Data")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()

# 線形回帰モデルのフィット
model = LinearRegression()
model.fit(x.reshape(-1, 1), y)

# 回帰直線
x_fit = np.linspace(min(x), max(x), 100)
y_fit = model.predict(x_fit[:, np.newaxis])

# 回帰結果をプロット
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Input data')
plt.plot(x_fit, y_fit, color='red', linewidth=2, label='Fit line')
plt.title("Linear Regression")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
